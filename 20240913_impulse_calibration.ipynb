```python
def get_unnormalized_amps(data_files, noise=False):
    amps = []
    for file in data_files:
        dtt, tt, nn = utils.load_timestreams(file, ['D', 'G'])
        fs = int(np.ceil(1/dtt))

        zz, dd = nn[0], nn[1]
        zz_bp = utils.bandpass_filtered(zz, fs, 40000, 130000)
        pulse_idx = utils.get_pulse_idx(dd, -0.5, False)
        
        if noise:
            pulse_idx = np.ceil(0.5 * (pulse_idx[:-1] + pulse_idx[1:])).astype(np.int64)

        for i, idx in enumerate(pulse_idx):
            if idx < 100000 or idx > tt.size-100000:
                continue
            window, f, f_lp, amp = utils.recon_pulse(idx, dtt, tt, zz_bp, dd, False, None)
            
            if noise:
                amps.append(np.abs(f_lp[np.ceil(f_lp.size/2+20).astype(np.int64)])/1e9)
            else:
                amps.append(amp)

    amps = np.asarray(amps)
    return amps

def gauss(x, A, mu, sigma):
    return A*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
folder = r'/Users/yuhan/work/nanospheres/data/20240913_two_e_data'

unnormalized_amps = []
datasets = ['20240913_1v_3e-8mbar', '20240913_2_5v_3e-8mbar', '20240913_5v_3e-8mbar', '20240913_10v_3e-8mbar']
pulseamps = [400, 1000, 2000, 4000]
for i, dataset in enumerate(datasets):
    print(dataset)
    combined_path = os.path.join(folder, dataset, '**/*.mat')
    data_files = glob.glob(combined_path)
    
    unnormalized_amps.append(get_unnormalized_amps(data_files, False))
```

```python
norm_factor = 4000 / np.mean(unnormalized_amps[3])
normalized_amps = [amp *norm_factor for amp in unnormalized_amps]
```

```python
hhs, bcs, gps = [], [], []
for amp in normalized_amps:
    bins = np.linspace(0, np.max(amp)*1.5, 50)
    hh, be = np.histogram(amp, bins=bins)
    bc = 0.5 * (be[1:] + be[:-1])
    
    gp, gcov = curve_fit(gauss, bc, hh, p0=[np.max(hh), np.mean(np.abs(amp)), np.std(np.abs(amp))])
    
    hhs.append(hh)
    bcs.append(bc)
    gps.append(gp)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
xx = np.linspace(0, np.max(np.asarray(bcs).flatten()), 1000)

for i, _ in enumerate(normalized_amps):
    ax.errorbar(bcs[i], hhs[i], yerr=np.sqrt(hhs[i]), fmt='o', color=yale_colors[i])
    ax.plot(xx, gauss(xx, *gps[i]), label=f'{pulseamps[i]} keV, $\sigma$ = {gps[i][2]:.1f} keV', color=yale_colors[i])

ax.set_xlabel('Reconstruced pulse (keV/c)', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.legend(fontsize=14)
```
