Take a look at windows being cut

```python
colors = plt.colormaps.get_cmap('tab20b').resampled(4).colors
pressures = ['3e-5', '3e-6', '5e-7', '2e-7']

fig, ax = plt.subplots(figsize=(8, 5))

for i, hists in enumerate([hists_0, hists_1, hists_2, hists_3]):
    hhs, hh_cut_det, hh_cut_all = hists[1], hists[2], hists[3]
    hh_cut_all_sum = np.sum(hh_cut_all, axis=0)

    n_search_per_win = (5000 - 150) / 25
    time_per_search = 2e-6 * 25
    scaling = n_search_per_win * time_per_search * (hists[0][1] - hists[0][0])
    time_integrated = hh_cut_all.shape[0] * n_search_per_win * time_per_search / 60

    ax.plot(hists[0], hh_cut_all_sum/(hh_cut_all.shape[0] * scaling), '-', color=colors[i], label=f'{pressures[i]} mbar ({time_integrated:.1f} minutes)')
    ax.legend(frameon=False, fontsize=12)

ax.set_yscale('log')
# ax.set_xscale('log')
ax.set_xlim(0, 6000)
ax.set_ylim(5e-7, 100)

ax.set_xlabel('Reconstructed amplitude (keV)')
ax.set_ylabel('Differential count (Hz/keV)')
```
