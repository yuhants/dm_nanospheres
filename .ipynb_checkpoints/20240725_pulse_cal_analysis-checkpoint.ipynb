```python
def get_recon_peaks(amps, tt, height=1.5, distance=3000):
    real_amp = np.real(amps)
    
    _all_peaks = []
    for i, amp in enumerate(real_amp):
        peaks, _ = find_peaks(amp, height=height, distance=distance)
        
        # Manually excluding first and last 0.25 seconds
        # because there seems to be bad reconstruction
        good_peaks = peaks[np.logical_and(tt[peaks] > 0.25, tt[peaks]<1.75)]
        _all_peaks.append(amp[good_peaks])

    recon_peaks = np.hstack(_all_peaks)
    
    return recon_peaks

def fit_recon_peaks(recon_peaks, center=30):
    mu, std = norm.fit(np.real(recon_peaks))
    
    scale_factor = center / mu
    sigma = std * scale_factor  # MeV
    
    return sigma, scale_factor
```

```python
recon_peaks_150v = get_recon_peaks(amp_150v/1e9, tt, height=1.6)
sig_150v, scale_150v = fit_recon_peaks(recon_peaks_150v, 30)
```

```python
recon_peaks_100v = get_recon_peaks(amp_100v/1e9, tt, height=1.0)
sig_100v, scale_100v = fit_recon_peaks(recon_peaks_100v, 20)
```

```python
recon_peaks_50v = get_recon_peaks(amp_50v/1e9, tt, height=0.5)
sig_50v, scale_50v = fit_recon_peaks(recon_peaks_50v, 10)
```

```python
xx150 = np.linspace(20, 40, 100)
pp150 = norm.pdf(xx150, 30, sig_150v)

xx100 = np.linspace(13, 27, 100)
pp100 = norm.pdf(xx100, 20, sig_100v)

xx50 = np.linspace(5, 15, 100)
pp50 = norm.pdf(xx50, 10, sig_50v)
```

```python
plt.figsize=(10, 4)

plt.xlim(0, 45)
plt.ylim(0, 0.5)
plt.hist(recon_peaks_150v*scale_150v, bins=20, density=True, alpha=0.6, color=colors[0])
plt.hist(recon_peaks_100v*scale_100v, bins=20, density=True, alpha=0.6, color=colors[1])
plt.hist(recon_peaks_50v*scale_50v, bins=20, density=True, alpha=0.6, color=colors[2])

plt.plot(xx150, pp150, '--', linewidth=1.5, color=colors[0])
plt.plot(xx100, pp100, '--', linewidth=1.5, color=colors[1])
plt.plot(xx50, pp50, '--', linewidth=1.5, color=colors[2])

plt.text(25, 0.2, f'$\sigma$ = {sig_150v:.2f} MeV/c', fontsize=12, color=colors[0])
plt.text(15, 0.25, f'$\sigma$ = {sig_100v:.2f} MeV/c', fontsize=12, color=colors[1])
plt.text(5, 0.45, f'$\sigma$ = {sig_50v:.2f} MeV/c', fontsize=12, color=colors[2])

plt.xlabel('Reconstructed pulse (MeV/c)')
plt.ylabel('Differential count')
```
